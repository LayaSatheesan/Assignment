```python
import pandas as pd
import numpy as ny
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read excel file from system

data_new=pd.read_excel("C:\\Users\\LAYA  S\\iris.xls")


#display the contents of dataset.It also display all the columns.

data_new
```

```python
#Calculate the mean of each column of the dataset.
data_new.mean()
```

```python
#Check for the null values present in the dataset.
#It gives False as output.
#So there is no missing values.
data_new.isnull().values.any()
```

# Create boxplot

```python
plt.boxplot(data_new['SL'])
plt.show()
```

# Create Scatter plot

```python
plt.scatter(data_new["SL"],data_new['PL'])
plt.title("Scatter plot between SL and PL")
plt.xlabel("SL")
plt.ylabel("PL")
plt.show()
```

# Histogram using matplotlib

```python
plt.figure(figsize=(8,6))
plt.hist(data_new['SL'])
plt.title("SL")
plt.xlabel("SL")
plt.ylabel("Count")
plt.show()
```

# Histogram using seaborn

```python
plt.figure(figsize=(8,6))
sns.histplot(x=data_new['SL'])
plt.title("SL")
plt.show()
```

# Boxplot

```python
plt.figure(figsize=(8,6))
plt.boxplot(data_new['SL'])
plt.ylabel("SL")
plt.show()
```

# Boxplot using seaborn

```python
plt.figure(figsize=(8,6))
sns.boxplot(x=data_new['Classification'],y=data_new['PL'])
plt.title("Classification Vs PL")
plt.xticks(rotation=90)
plt.show()
```

# Scatterplot

```python
plt.scatter(data_new['SL'],data_new['SW'])
plt.title("SL Vs SW")
plt.xlabel("SL")
plt.ylabel("SW")
plt.show()
```

# Violin Plot

```python
sns.violinplot(x=data_new['Classification'],y=data_new['PL'])
plt.title("Classification Vs PL")
plt.show()
```

# Pairplot

```python
data_2=data_new[['SL','PL','SW','PW']]
sns.pairplot(data_2)
plt.show()
```

```python
sns.countplot(x=data_new['Classification'])
plt.xlabel("Classification")
plt.show()
```
